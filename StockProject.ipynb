```python
import FinanceDataReader as fdr

# 한국거래소(KRX)의 주식 목록 불러오기
df_krx = fdr.StockListing('KRX')

# 데이터프레임의 열 이름 출력
print(df_krx.columns)
```

```python
from datetime import datetime, timedelta


# 상위 10개 종목 선택 (시가총액 정보가 없으므로 상위 10개 종목을 임의로 선택)
top_10_stocks = df_krx.head(10)

# 최근 2년간의 데이터를 가져오기 위한 날짜 설정
end_date = datetime.now()
start_date = end_date - timedelta(days=2*365)

# 각 종목에 대한 최근 2년간의 주가 데이터 수집
stock_data = {}
for idx, row in top_10_stocks.iterrows():
    code = row['Code']  # 'Symbol' 대신 'Code' 사용
    stock_data[code] = fdr.DataReader(code, start_date, end_date)

# 첫 번째 종목의 데이터를 예시로 출력
print(stock_data[top_10_stocks.iloc[0]['Code']].head())
```
